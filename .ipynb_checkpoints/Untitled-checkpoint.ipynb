```python
import cv2 
import time
import os
import uuid
```

```python
IMAGES_PATH = 'Tensorflow/workspace/images/collectedImages'
```

```python
labels = ['hello', 'yes', 'no', 'iloveyou']
img_count = 15
```

```python
for label in labels:
    os.makedirs(os.path.join('Tensorflow', 'workspace', 'images', 'collectedImages', label), exist_ok=True)
    cap = cv2.VideoCapture(0)
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    for img_num in range (img_count):
        ret, frame = cap.read()
        img_name = os.path.join(IMAGES_PATH, label, f"{label}.{uuid.uuid1()}.jpg")
        cv2.imwrite(img_name, frame)
        cv2.imshow('frame', frame)
        time.sleep(2)

        if cv2.waitKey(1) & 0xff == ord('q'):
            break
    cap.release()
```
